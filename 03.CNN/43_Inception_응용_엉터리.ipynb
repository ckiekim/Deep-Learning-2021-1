# GoogLeNet(Inception v3) 응용

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3, decode_predictions
```

```python
inception = InceptionV3()
inception.summary()
```

```python
import numpy as np
import cv2

img = cv2.imread('images/고양이.jpg', -1)
img = cv2.resize(img, (299, 299))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)      # img = img[:,:,::-1]
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
yhat = inception.predict(img.reshape(-1, 299, 299, 3))
label = decode_predictions(yhat)
label = label[0][0]
print(f'{label[1]} ({label[2]*100:.2f}%)')
```

```python
img = cv2.imread('images/상어.jpg', -1)
img = cv2.resize(img, (299, 299))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis('off')
plt.show()

yhat = inception.predict(img.reshape(-1, 299, 299, 3))
label = decode_predictions(yhat)
label = label[0][0]
print(f'{label[1]} ({label[2]*100:.2f}%)')
```
